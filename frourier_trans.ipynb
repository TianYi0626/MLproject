```python
import pandas as pd
import numpy as np

X_train = pd.read_csv('8000rpm_data_z-score/train_sample.csv').values  # Assuming this is a 2D array

X_train_fft = np.fft.fft(X_train, axis=1)  # Perform FFT along each column (axis=0)

X_train_fft_magnitude = np.abs(X_train_fft)

# Step 4: Save the transformed data to a new CSV file
pd.DataFrame(X_train_fft_magnitude).to_csv('8000rpm_data_z-score/train_sample_fft.csv', index=False)
```

```python
X_test = pd.read_csv('8000rpm_data_z-score/test_sample.csv').values  # Assuming this is a 2D array

X_test_fft = np.fft.fft(X_test, axis=1)  # Perform FFT along each column (axis=0)

X_test_fft_magnitude = np.abs(X_test_fft)

# Step 4: Save the transformed data to a new CSV file
pd.DataFrame(X_test_fft_magnitude).to_csv('8000rpm_data_z-score/test_sample_fft.csv', index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Plot the original signal (first row as an example)
plt.subplot(2, 1, 1)
plt.plot(X_train[0, :])
plt.title('Original Signal (First Row) in Time Domain')
plt.xlabel('Samples')
plt.ylabel('Amplitude')

# Plot the magnitude of the Fourier Transform (frequency domain)
plt.subplot(2, 1, 2)
plt.plot(np.fft.fftfreq(X_train.shape[1]), X_train_fft_magnitude[0, :])  # Frequency axis
plt.title('Magnitude of FFT (Frequency Domain) - First Column')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Plot the original signal (first column as an example)
plt.subplot(2, 1, 1)
plt.plot(X_test[0, :])
plt.title('Original Signal (First Row) in Time Domain')
plt.xlabel('Samples')
plt.ylabel('Amplitude')

# Plot the magnitude of the Fourier Transform (frequency domain)
plt.subplot(2, 1, 2)
plt.plot(np.fft.fftfreq(X_test.shape[0]), X_test_fft_magnitude[:, 0])  # Frequency axis
plt.title('Magnitude of FFT (Frequency Domain) - First Column')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```
